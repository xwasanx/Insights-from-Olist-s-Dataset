```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.express as px
```

```python
order1=pd.read_csv('olist_order_items_dataset_2.csv')
order1
```

```python
order2=pd.read_csv('orders2.csv')
order2
```

```python
join_order=order2.merge(order1,on='order_id')
join_order
```

```python
join_order['full_price']=join_order.price + join_order.freight_value
```

```python
join_order
```

```python
join_order['delay']=pd.to_datetime(join_order['order_delivered_customer_date']).dt.date - pd.to_datetime(join_order['order_estimated_delivery_date']).dt.date
```

```python
join_order
```

```python
order_status_counts = join_order['order_status'].value_counts()
plt.title('Order Status Distribution')
plt.bar(order_status_counts.index, order_status_counts.values)
plt.xlabel('Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
join_order['order_delivered_customer_date'] = pd.to_datetime(join_order['order_delivered_customer_date'])
join_order['order_estimated_delivery_date'] = pd.to_datetime(join_order['order_estimated_delivery_date'])

join_order['delivery_delay'] = (join_order['order_delivered_customer_date'] - join_order['order_estimated_delivery_date']).dt.days

daily_delivery_delay = join_order.groupby(join_order['order_delivered_customer_date'].dt.date)['delivery_delay'].mean()

plt.figure(figsize=(10, 6))
plt.plot(daily_delivery_delay.index, daily_delivery_delay.values)
plt.xlabel('Date')
plt.ylabel('Delivery Delay (Days)')
plt.title('Average Delivery Delay Over Time')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(join_order['price'], join_order['freight_value'])
plt.xlabel('Price')
plt.ylabel('Freight Value')
plt.title('Price vs. Freight Value')
plt.show()
```

1. Delivery Performance: The increasing average delivery delay over time highlights the need to address delivery inefficiencies and improve overall delivery performance. This can help enhance customer satisfaction and loyalty.

2.Exploring pricing adjustments and optimizing shipping costs may help improve profitability while meeting customer expectations.
